```python
#Natural Join
```

```python
connection=sql.connect(host='localhost',user='root',password="rana1998@",database="asig_1_sql_")
cursor=connection.cursor()
cursor.execute("SELECT * FROM customer JOIN orders ON customer.id =orders.customer_id")
items=cursor.fetchall()

df=pd.DataFrame(items,columns=['id(PK)','name','age','address','salary','oid(PK)','date','customer_id','amount '])
df
```

```python
#Left Join
```

```python
connection=sql.connect(host='localhost',user='root',password="rana1998@",database="asig_1_sql_")
cursor=connection.cursor()
cursor.execute("SELECT * FROM customer LEFT JOIN orders ON customer.id =orders.customer_id")
items=cursor.fetchall()

df=pd.DataFrame(items,columns=['id(PK)','name','age','address','salary','oid(PK)','date','customer_id','amount '])
df
```

```python
#Right Join
```

```python
connection=sql.connect(host='localhost',user='root',password="rana1998@",database="asig_1_sql_")
cursor=connection.cursor()
cursor.execute("SELECT * FROM customer RIGHT JOIN orders ON customer.id =orders.customer_id")
items=cursor.fetchall()

df=pd.DataFrame(items,columns=['oid(PK)','date','customer_id','amount','id(PK)','name','age','address','salary'])
df
```

```python
#Full Outer Join
```

```python
#FULL Outer Join is the uinion of Left Join and Right Join
connection=sql.connect(host='localhost',user='root',password="rana1998@",database="asig_1_sql_")
cursor=connection.cursor()
cursor.execute("""
SELECT * FROM customer LEFT JOIN orders ON customer.id =orders.customer_id
UNION
SELECT * FROM customer RIGHT JOIN orders ON customer.id =orders.customer_id
""") 
items=cursor.fetchall()

df=pd.DataFrame(items,columns=['id(PK)','name','age','address','salary','oid(PK)','date','customer_id','amount'])
df
```
