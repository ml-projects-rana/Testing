```python
import sqlite3
import pandas as pd
```

```python
#Eiher connect to the given database(idf exist) or created new database with the given name
conn=sqlite3.connect('sqlite_database.db')
```

```python
#create cursor
c=conn.cursor()
```

```python
#create customer table
c.execute("""
            CREATE TABLE customer(
            id int NOT NULL,
            name text,
            age int,
            address text,
            salary int,
            PRIMARY KEY(id))
       """)
conn.commit()
```

### Insertion

```python
#create orders table
c.execute("""
CREATE TABLE orders (
            oid int NOT NULL,
            Date date,
            customer_id int,
            amount int,
            PRIMARY KEY(oid))
       """)
conn.commit()
```

### Delete operation

```python
#connecting to database
conn=sqlite3.connect('sqlite_database.db')
#create cursor
c=conn.cursor()
c.execute("DELETE FROM customer where id=7")
conn.commit() #Need to commit otherwise delete operation will not proceed

c.execute("SELECT * FROM customer ")
items=c.fetchall()
df=pd.DataFrame(items,columns=['id','name','age','address','salary'])
conn.close()
df
```

### Update operation

```python
conn=sqlite3.connect('sqlite_database.db')
#create cursor
c=conn.cursor()
c.execute("UPDATE customer SET age=29 WHERE id=6")
conn.commit() #Need to commit otherwise update operation will not proceed
c.execute("SELECT * FROM customer ")
items=c.fetchall()
df=pd.DataFrame(items,columns=['id','name','age','address','salary'])
conn.close()
df
```

### Natural Join

```python
conn=sqlite3.connect('sqlite_database.db')
#create cursor
c=conn.cursor()

c.execute("SELECT * FROM customer JOIN orders ON customer.id =orders.customer_id")
items=c.fetchall()

df=pd.DataFrame(items,columns=['id(PK)','name','age','address','salary','oid(PK)','date','customer_id','amount '])
conn.close()
df
```

### Left Join

```python
conn=sqlite3.connect('sqlite_database.db')
#create cursor
c=conn.cursor()

c.execute("SELECT * FROM customer LEFT JOIN orders ON customer.id =orders.customer_id")
items=c.fetchall()

df=pd.DataFrame(items,columns=['id(PK)','name','age','address','salary','oid(PK)','date','customer_id','amount '])
conn.close()
df
```

### Right Join

```python
conn=sqlite3.connect('sqlite_database.db')
#create cursor
c=conn.cursor()

c.execute("SELECT * FROM orders LEFT JOIN customer ON orders.customer_id=customer.id")
items=c.fetchall()
df=pd.DataFrame(items,columns=['oid(PK)','date','customer_id','amount','id(PK)','name','age','address','salary'])
conn.close()
df
```
